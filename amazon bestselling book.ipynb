```python
corr = data.corr(numeric_only=True)

plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
# Insight 1
# Books with more weeks on the bestseller list tend to receive more customer reviews (correlation = 0.55).
# Insight 2
# Ratings and reviews have only a weak relationship (correlation = 0.15), suggesting popularity does not always imply higher ratings.
# Insight 3
# Book price shows almost no correlation with ratings or reviews, indicating that expensive books are not necessarily more popular.
# Insight 4
# Higher-ranked books tend to have more reviews, showing a connection between popularity and ranking.
```
